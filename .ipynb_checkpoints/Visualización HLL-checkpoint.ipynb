```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
%matplotlib inline
```

```python
data_lax = np.loadtxt('euler_lax-04.dat')
```

```python
data_hll = np.loadtxt('../hll2/euler_lax-02.dat')
```

```python
os.listdir()
```

```python
plt.figure(figsize=(25,16))


plt.subplot(2,2,1)
#plt.plot(data[:,0], data[:,1], 'r--')
plt.scatter(data_lax[:,0], data_lax[:,1],marker='*')
plt.scatter(data_hll[:,0], data_hll[:,1],marker='*')
plt.xlabel('Position', fontsize=20)
plt.ylabel('Density', fontsize=20)
plt.grid(True)

plt.subplot(2,2,2)
plt.scatter(data_lax[:,0], data_lax[:,2] )
plt.scatter(data_hll[:,0], data_hll[:,2])
plt.xlabel('Position', fontsize=20)
plt.ylabel('Velocity', fontsize=20)
plt.ylim(0,1.6)
plt.grid(True)

plt.subplot(2,2,3)
plt.scatter(data_lax[:,0], data_lax[:,3])
plt.scatter(data_hll[:,0], data_hll[:,3])
plt.xlabel('Position', fontsize=20)
plt.ylabel('Pressure', fontsize=20)
plt.grid(True)
```
