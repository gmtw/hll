```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
%matplotlib inline
```

```python
def Ploteo(data_lax, data_hll,y11=0.0,y21=1.1,y12=0.0,y22=1.6,y13=0.0,y23=1.1 ):
    plt.figure(figsize=(25,16))


    plt.subplot(2,2,1)
    #plt.plot(data[:,0], data[:,1], 'r--')
    plt.scatter(data_lax[:,0], data_lax[:,1],marker='*', label = 'Lax-Friederich')
    plt.scatter(data_hll[:,0], data_hll[:,1],marker='.', label = 'HLL')
    plt.xlabel('Position', fontsize=20)
    plt.ylabel('Density', fontsize=20)
    plt.ylim(y11,y21)
    plt.legend()
    plt.grid(True)

    plt.subplot(2,2,2)
    plt.scatter(data_lax[:,0], data_lax[:,2] ,marker='v', label = 'Lax-Friederich')
    plt.scatter(data_hll[:,0], data_hll[:,2] ,marker='o', label = 'HLL')
    plt.xlabel('Position', fontsize=20)
    plt.ylabel('Velocity', fontsize=20)
    plt.ylim(y12,y22)
    plt.legend()
    plt.grid(True)

    plt.subplot(2,2,3)
    plt.scatter(data_lax[:,0], data_lax[:,3],marker='1', label = 'Lax-Friederich')
    plt.scatter(data_hll[:,0], data_hll[:,3],marker='P', label = 'HLL')
    plt.xlabel('Position', fontsize=20)
    plt.ylabel('Pressure', fontsize=20)
    plt.ylim(y13,y23)
    plt.legend()
    plt.grid(True)
```

# Test 1

Para $t_{LF}=0.2$

Para $t_{HLL}=0.1$

```python
data_hll = np.loadtxt('../hll2/euler_lax-02.dat')
data_lax = np.loadtxt('euler_lax-04.dat')
```

```python
Ploteo(data_lax, data_hll, y12=-0.1)
```
